```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
import string

nltk.download('stopwords')

stemmer = nltk.SnowballStemmer('english')
stopword = set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()                                          # transform text to lowercase
    text = re.sub('\[.*?\]', '', text)                                # remove '[]' and words in it
    text = re.sub('https?://\S+|www\.\S+', '', text)                  # remove links
    text = re.sub('<.*?>+', '', text)                                 # remove '<>' and words in it
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)   # remove punctuation
    text = re.sub('\n', '', text)                                     # remove line feeds
    text = re.sub('\w*\d\w*', '', text)                               # remove numbers and words with numbers
    text = [word for word in text.split(' ') if word not in stopword] # remove stopwords
    text = [stemmer.stem(word) for word in text]                      # stem the text
    text=" ".join(text)                                               # put words back together to form a text
    return text

data['review_description_clean'] = data['review_description'].apply(clean)

data['review_description_clean'].head()
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sentiments = SentimentIntensityAnalyzer()

data['positive'] = [sentiments.polarity_scores(i)['pos'] for i in data['review_description_clean']]
data['negative'] = [sentiments.polarity_scores(i)['neg'] for i in data['review_description_clean']]
data['neutral'] = [sentiments.polarity_scores(i)['neu'] for i in data['review_description_clean']]
data['compound'] = [sentiments.polarity_scores(i)['compound'] for i in data['review_description_clean']]
data = data[['review_description', 'review_description_clean', 'rating', 'positive', 'negative', 'neutral', 'compound']]
data
```

```python
result = data["compound"].values
sentiment = []

for i in result:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
        
data["sentiment"] = sentiment
data.head()
```

```python
scores = data['sentiment'].value_counts()

plt.pie(data['sentiment'].value_counts(), 
        labels=scores.index, 
        autopct='%1.1f%%', 
        radius=1, 
        colors=['#3bcc44','#ff3313', '#ffd420'])
plt.show()
```

```python
df_positive = data.loc[data['sentiment'] == 'Positive']
df_negative = data.loc[data['sentiment'] == 'Negative']
text_positive = ' '.join(i for i in df_positive['review_description_clean'])
text_negative = ' '.join(i for i in df_negative['review_description_clean'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('omw-1.4')

docs = [text_positive, text_negative]

cv = CountVectorizer()
word_count = cv.fit_transform(docs)
tf = pd.DataFrame(word_count.toarray(), columns=cv.get_feature_names_out())
tfidf_transformer = TfidfTransformer()
x = tfidf_transformer.fit_transform(word_count)
tfidf = pd.DataFrame(x.toarray() ,columns=cv.get_feature_names_out())
tfidf = tfidf.transpose().rename(columns={0:"positive", 1:"negative"}).reset_index()

tfidf
```

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')

tfidf['pos'] = pos_tag(tfidf['index'])
tfidf['index'], tfidf['pos'] = tfidf['pos'].str
tfidf_nouns= tfidf.loc[(tfidf['pos'] == 'NN') | (tfidf['pos'] == 'NNS') | (tfidf['pos'] == 'NNP') | (tfidf['pos'] == 'NNPS'), ['index', 'positive', 'negative', 'pos']]
tfidf_adjectives= tfidf.loc[(tfidf['pos'] == 'JJ') | (tfidf['pos'] == 'JJR') | (tfidf['pos'] == 'JJS'), ['index', 'positive', 'negative', 'pos']]
tfidf_nouns = pd.melt(tfidf_nouns, id_vars=['index', 'pos'], value_vars=['positive', 'negative']).sort_values('value', ascending=False)
tfidf_adjectives = pd.melt(tfidf_adjectives, id_vars=['index', 'pos'], value_vars=['positive', 'negative']).sort_values('value', ascending=False)
tfidf_plot = pd.concat([tfidf_nouns, tfidf_adjectives])
tfidf_plot = tfidf_plot[tfidf_plot['value'] != 0]

tfidf_plot
```

```python
from ipywidgets import widgets, interactive
import plotly.express as px

# sentiment dropdown
sentiment_dropdown = widgets.Dropdown(options=['all'] + list(tfidf_plot['variable'].unique()),
                                      value='all',
                                      description='Sentiment:')
# parts of speech dropdown
pos_dropdown = widgets.Dropdown(options=['all', 'nouns', 'adjectives'],
                                      value='all',
                                      description='Part of speech:')

# function to plot charts
def plot_charts(sentiment_dropdown, pos_dropdown):
    df_temp = tfidf_plot.copy()
    if sentiment_dropdown != 'all':
        df_temp = df_temp[df_temp['variable'] == sentiment_dropdown]
    if pos_dropdown == 'nouns':
        df_temp = df_temp[df_temp['pos'].isin(['NN', 'NNS', 'NNP', 'NNPS'])]
    elif pos_dropdown == 'adjectives':
        df_temp = df_temp[df_temp['pos'].isin(['JJ', 'JJR', 'JJS'])]
            
    df_temp.groupby(['index', 'variable', 'value']).sum().unstack()
    
    # treemap plot 
    figure = px.treemap(df_temp[0:30],
                        path=['index'],
                        values='value',
                        color='value',
                        hover_data=['variable'],
                        color_continuous_scale=colors[::-1])    
    figure.show()

interactive(plot_charts, sentiment_dropdown=sentiment_dropdown, pos_dropdown=pos_dropdown)
```

```python
import plotly.graph_objects as go

# word to be searched
word_pick = widgets.Combobox(placeholder='example: match',
                             options=list(tfidf_plot['index'].unique()),
                             description='Choose word:')

# function to plot table
def plot_charts(word_pick):
    df_temp = data.copy()
    df_temp = df_temp[df_temp['review_description'].str.contains(word_pick)]      
    df_temp.groupby(['review_description', 'rating', 'sentiment']).sum().unstack()
    
    # table plot 
    table = go.Figure(data=[go.Table(header=dict(values=['original review','rating','sentiment'],
                                               fill_color='paleturquoise',
                                               align='left'),
                                   cells=dict(values=[df_temp['review_description'][:10], df_temp['rating'][:10], df_temp['sentiment'][:10]],
                                              fill_color='lavender',
                                              align='left'), columnwidth=[10,1,1])])

    table.show()

interactive(plot_charts, word_pick=word_pick)
```
